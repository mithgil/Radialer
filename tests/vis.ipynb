## Importing txt file

```python
from read import read_4d_complex_array_from_file
import numpy as np

filename = '../build/radial_array.txt'
dim1, dim2, dim3, dim4 = 3, 81, 81, 81 
complex_array = read_4d_complex_array_from_file(filename, dim1, dim2, dim3, dim4)
print("Shape of the loaded array:", complex_array.shape)
```

```python
complex_array[1,1,1,40]
```

```python

slicing = 40
xsliced = complex_array[0,:,:,slicing]
ysliced = complex_array[1,:,:,slicing]
zsliced = complex_array[2,:,:,slicing]

import matplotlib.pyplot as plt


fig, ax = plt.subplots(2,2, figsize =(10, 10))

ax[0,0].imshow(np.square(abs(xsliced)))
ax[0,0].set_title(r'|Ex|$^{2}$')
ax[0,1].imshow(np.square(abs(ysliced)))
ax[0,1].set_title(r'|Ey|$^{2}$')
ax[1,0].imshow(np.square(abs(xsliced))+np.square(abs(ysliced)))
ax[1,0].set_title(r'|Ex|$^{2}$+|Ey|$^{2}$')
ax[1,1].imshow(np.square(abs(zsliced)))
ax[1,1].set_title(r'|Ez|$^{2}$')

plt.tight_layout()

plt.show()
```

## Importing npy file

```python
import numpy as np

filename = '../build/radial_array.npy'
rad_arr = np.load(filename)
```

```python
rad_arr.shape
```

```python

slicing = 40
xsliced = rad_arr[0,:,:,slicing]
ysliced = rad_arr[1,:,:,slicing]
zsliced = rad_arr[2,:,:,slicing]

import matplotlib.pyplot as plt


fig, ax = plt.subplots(2,2, figsize =(10, 10))

ax[0,0].imshow(np.square(xsliced.real))
ax[0,0].set_title(r'|Ex|$^{2}$')
ax[0,1].imshow(np.square(ysliced.real))
ax[0,1].set_title(r'|Ey|$^{2}$')
ax[1,0].imshow(np.square(xsliced.real)+np.square(ysliced.real))
ax[1,0].set_title(r'|Ex|$^{2}$+|Ey|$^{2}$')
ax[1,1].imshow(np.square(zsliced.real))
ax[1,1].set_title(r'|Ez|$^{2}$')

plt.tight_layout()
plt.savefig("radial_mode_component_xy_intensity.png", bbox_inches = 'tight',dpi=500)

plt.show()
```

```python

slicing = 40
x2sliced = rad_arr[0,:,slicing,:].T
y2sliced = rad_arr[1,:,slicing,:].T
z2sliced = rad_arr[2,:,slicing,:].T

import matplotlib.pyplot as plt


fig, ax = plt.subplots(2,2, figsize =(10, 10))

ax[0,0].imshow(np.square(abs(x2sliced)))
ax[0,0].set_title(r'|Ex|$^{2}$')
ax[0,1].imshow(np.square(abs(y2sliced)))
ax[0,1].set_title(r'|Ey|$^{2}$')
ax[1,0].imshow(np.square(abs(x2sliced))+np.square(abs(y2sliced)))
ax[1,0].set_title(r'|Ex|$^{2}$+|Ey|$^{2}$')
ax[1,1].imshow(np.square(abs(z2sliced)))
ax[1,1].set_title(r'|Ez|$^{2}$')

plt.tight_layout()
plt.savefig("radial_mode_component_xz_intensity.png", bbox_inches = 'tight',dpi=500)

plt.show()
```

```python

hwindow = 1 # um, from c++ user 

x = np.linspace(-hwindow, hwindow, 81)
y = np.linspace(-hwindow, hwindow, 81)

X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(figsize =(14, 8))

ax.quiver(X, Y, xsliced.real, ysliced.real)

ax.set_aspect('equal')
plt.savefig("radial_mode_exy_vector.png", bbox_inches = 'tight',dpi=500)

plt.show()
```
